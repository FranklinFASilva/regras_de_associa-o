# Resultado da pesquisa de acesso à redes sociais

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

## Calculando o suporte

```python
dados2 = pd.read_csv('novo.csv') #lendo o csv
```

```python
#calculando o suporte
suportes2 = apriori(dados2, min_support=0.6, use_colnames=True)
suportes2 #imprime os suportes
```

## Gerando regras de Associação

```python
#mandando calcular as regras com lift de 1.218
from mlxtend.frequent_patterns import association_rules
regras2 = association_rules(suportes2, metric="confidence", min_threshold=0.8)
regras2
```

```python
regras2.to_excel('novo1.xlsx',encoding='utf-8',index=False)
```
